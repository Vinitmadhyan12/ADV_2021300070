```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset


# Load the CSV file into a DataFrame
df = pd.read_csv('data.csv')

# Display the first few rows of the dataframe
df.head()




category_sales = df.groupby('Product Category')['Total Revenue'].sum().sort_values()

# Plot the bar chart for Product Category wise sales
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Product Category Wise Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()


# 2. Pie chart: Region wise sales
region_sales = df.groupby('Region')['Total Revenue'].sum()
plt.figure(figsize=(8, 8))
region_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Region Wise Sales')
plt.ylabel('')
plt.show()

# 3. Histogram: Units Sold
plt.figure(figsize=(10, 6))
plt.hist(df['Units Sold'], bins=20, edgecolor='k')
plt.title('Histogram of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()

# 4. Timeline chart: Sales over time
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df_resampled = df.resample('M').sum()
plt.figure(figsize=(12, 6))
plt.plot(df_resampled.index, df_resampled['Total Revenue'], marker='o')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()

# 5. Scatter plot: Units Sold vs. Total Revenue
plt.figure(figsize=(10, 6))
plt.scatter(df['Unit Price'], df['Units Sold'])
plt.title('Units Sold vs. Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Units Sold')
plt.show()

# Bubble Plot: Total Revenue vs. Units Sold with Bubble Size as Unit Price
plt.figure(figsize=(12, 8))
plt.scatter(df['Units Sold'], df['Total Revenue'], s=df['Unit Price'] * 10, alpha=0.5, c='blue')
plt.title('Total Revenue vs. Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Total Revenue')
plt.show()
```
